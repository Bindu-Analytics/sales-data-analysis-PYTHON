## Task 1 – <span style="color:blue">Create DataFrames and Save as CSV Files</span>  



```python
import pandas as pd
#Importing pandas library
import numpy as np 
#Importing numpy library

# Assign Excel file paths to separate variables
project_file = 'Project.xlsx'
employees_file = 'Employees.xlsx'
seniority_file = 'Seniority.xlsx'


# Read each Excel file into its own DataFrame
Project_df = pd.read_excel(project_file)
Employees_df = pd.read_excel(employees_file)
Seniority_df = pd.read_excel(seniority_file)

# Save each DataFrame as a CSV file
Project_df.to_csv('project.csv', index=False)
Employees_df.to_csv('employees.csv', index=False)
Seniority_df.to_csv('seniority.csv', index=False)

#  Display DataFrames

from IPython.display import display

display(Project_df)
display(Employees_df)
display(Seniority_df)
```

## Task 2 –  <span style="color:sblue">Replace Missing Cost Values with Running Average (Using for Loop)</span>  

```python
#Read the CSV file 
project_df = pd.read_csv('project.csv')
# Fill missing values in 'Cost' using running average
total = 0 
count = 0 

for i in range(len(project_df)):
  # If cost value is missing (NaN)
    if pd.isnull(project_df.loc[i, 'Cost']):
        if count > 0:
            avg = total / count   # running average
            project_df.loc[i, 'Cost'] = avg  # replace NaN with running average
        else:
            project_df.loc[i, 'Cost'] = 0    # first NaN, set to 0
    else:
        # Add cost to total and increment count
        total += project_df.loc[i, 'Cost']
        count += 1

# display result 
pd.set_option('display.float_format', '{:.0f}'.format)

# save to CSV for further tasks
project_df.to_csv('project.csv', index=False)

project_df
```

## Task 3 – <span style="color:blue">Split Name Column into First and Last Name</span>  

```python
# Read the Project data from your CSV file
Employees_df = pd.read_csv('employees.csv')

# Only do split if the 'Name' column exists
if 'Name' in Employees_df.columns:
    # Split 'Name' column into two columns: 'First Name' and 'Last Name'
    Employees_df[['First Name', 'Last Name']] = Employees_df['Name'].str.split(' ', n=1, expand=True)
    # Remove the original 'Name' column from the dataframe
    Employees_df = Employees_df.drop(columns=['Name'])
    # Save the updated dataframe to CSV for further use 
    Employees_df.to_csv('employees_updated.csv', index=False)

# To verify and display output 
Employees_df
```

## Task 4 – <span style="color:blue">Join All Three DataFrames into One Final DataFrame Named "FINAL"</span>  

```python
# Load the updated employee CSV with changes made 
Employees_df = pd.read_csv('employees_updated.csv')  

# Load the latest project CSV file
project_df = pd.read_csv('project.csv')  

# Load the seniority CSV file 
Seniority_df = pd.read_csv('Seniority.csv')

# Merge the project and employee dataframes on 'ID', keeping all rows from project_df
merged = pd.merge(project_df, Employees_df, on='ID', how='left')

# Merge the above result with designation dataframe on 'ID', again keeping all rows
Final = pd.merge(merged, Seniority_df, on='ID', how='left')

# Remove any rows with missing values and reset the row index for a clean result
Final = Final.dropna().reset_index(drop=True)

# To Verify and display updated merged dataframe
Final
```

## Task 5 – <span style="color:blue">Add Bonus (5% of Cost for Finished Projects)</span>  

```python
# For each row in the dataframe, if the project is 'Finished', calculate Bonus as 5% of Cost, else set Bonus to 0
Final['Bonus'] = Final.apply(
    lambda row: row['Cost'] * 0.05 if str(row['Status']).lower() == 'finished' else 0,
    axis=1 )
#Mention axis and required feilds     
# Verify display the output
Final
```

## Task 6 – <span style="color:blue">Demote Designation Level by 1 for Failed Projects, Drop If > 4</span>  

```python
## Decrease the 'Designation Level' by 1 for all records where the 'Status' column value is 'fail' 
Final.loc[Final['Status'].str.lower() == 'fail', 'Designation Level'] -= 1

# Filter the DataFrame to include only rows where 'Designation Level' is less than or equal to 4, 
# and reset the index to maintain a clean sequential order
# Delete records where designation level is now greater than 4
Final = Final[Final['Designation Level'] <= 4].reset_index(drop=True)

# Display the output
Final
```

## Task 7 – <span style="color:blue"> Add Mr./Mrs. Prefix and Drop Gender Column</span>  

```python
# First, check if both 'Gender' and 'First Name' columns exist in the DataFrame
if 'Gender' in Final.columns and 'First Name' in Final.columns:
    # For each row, add "Mr." if gender is Male ("M"), otherwise add "Mrs." before the First Name value
    Final['First Name'] = Final.apply(
        lambda row: ('Mr. ' if str(row['Gender']).upper() == 'M' else 'Mrs. ') + str(row['First Name']),
        axis=1
    )
    # After prefixing, remove the 'Gender' column from the dataframe
    Final = Final.drop(columns=['Gender'])

# Display the Result after changes 
Final
```

## Task 8 – <span style="color:blue">Promote Designation by 1 for Employees Over Age 29</span>  

```python
# Increase the 'Designation Level' by 1 for all records where the 'Age' is greater than 29 using IF condition
Final.loc[Final['Age'] > 29, 'Designation Level'] += 1 

# Display updated dataframe
Final
```

## Task 9 – <span style="color:blue">Total Project Cost per Employee</span>  

```python
# To Calculate total project cost for each employee

# Group the DataFrame by 'ID' and 'First Name'
TotalProjectCost = Final.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

# Rename the 'Cost' column to 'TotalProjCost'
TotalProjectCost = TotalProjectCost.rename(columns={'Cost': 'TotalProjectCost'})

# Show  the resulting table as output
TotalProjectCost
```
